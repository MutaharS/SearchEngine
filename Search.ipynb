```python
"""
Now we need to implement a way to search through our content.json for the most relevant links
"""

import json

lookup_file = 'json_files/lookup_1gram.json'
with open(lookup_file, 'r') as lookup_json:
  lookup = json.load(lookup_json)

"""
Searching with 1gram (single word matching)
"""
# Given a query, gather the most relevant links for that search term
def search(query: str):
	# First we split the search string into each individual word
	words = query.strip().split(' ')
	
	results = []
	for word in words:
		links = lookup.get(word.lower(), None)
		if links:
			results.extend(links)
	return sorted(results, key=lambda record: record[1], reverse=True)

flag = 'Y'
while('y' in flag.lower()):
  qry = input('Search: ')
  results = search(qry)
  for i,res in enumerate(results):
    print(res)
    if i+1 % 3 == 0:
      break
  flag = input('Continue? : ')
  while('y' not in flag.lower() and not 'n' in flag.lower()):
    flag = input('Invalid option, continue? : ')
```

```python
"""
Now we need to implement a way to search through our content.json for the most relevant links
"""

import json

class NGramException(Exception):
  pass

def NGram(content, N=2):
  words = content.lower().split(' ')
  if N > len(words):
    raise NGramException('Trying to split into N-Grams on content with less than N words.')
  grams = []
  for i in range(len(words) - N + 1):
    grams.append(' '.join(words[i : i + N]))
  return grams

lookup_file = 'json_files/lookup_2gram.json'
with open(lookup_file, 'r') as lookup_json:
  lookup = json.load(lookup_json)
```

```python
"""
Searching with 2grams
"""

# Given a query, gather the most relevant links for that search term
def search(query: str):
  # First we split the search string into 2-Grams
  words = NGram(query, N=2)
  results = []
  for word in words:
    links = lookup.get(word.lower(), None)
    if links:
      results.extend(links)
  print(results)
  return sorted(results, key=lambda record: record[1], reverse=True)

flag = 'Y'
while('y' in flag.lower()):
  qry = input('Search: ')
  results = search(qry)
  for i,res in enumerate(results):
    print(res)
    if i+1 % 3 == 0:
      break
  flag = input('Continue? : ')
  while('y' not in flag.lower() and not 'n' in flag.lower()):
    flag = input('Invalid option, continue? : ')
```
